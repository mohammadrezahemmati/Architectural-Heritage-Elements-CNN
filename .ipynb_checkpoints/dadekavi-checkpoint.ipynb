```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
# A: Loading Dataset
df = pd.read_csv("marketing_campaign.csv")
```

```python
# C: Observe first five rows
df.head()
```

```python
# E: Checking Data integrity
```

```python
# F: Drop ID col
df = df.drop("ID", axis=1)
df.columns
```

```python
# G: Summary statistics
df.describe()
```

```python
# H: Explore in cathegorical variable
cathegorical_indices = [
  "Education", "Marital_Status", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5",
  "AcceptedCmp1", "AcceptedCmp2", "Complain", "Response"
]

for ind in cathegorical_indices:
  cathegories = list(set(df[ind]))
  print(ind, ":", cathegories, "with", len(cathegories), "cathegories.")
  for cat in cathegories:
    data_no = (df[ind] == cat).sum()
    print("\tCathegory", cat, "contatins", data_no, "datas.")
```

```python
# I: Replace “2n Cycle” with “Master” in “Education” column.
df["Education"] = df["Education"].replace(["Master"], "2n Cycle")
df["Education"]
```

```python
# J: Replace [“Alone”, “Absurd”, “YOLO”] with “Single” in “Martial_Status” column
df["Marital_Status"] = df["Marital_Status"].replace(["Alone", "Absurd", "YOLO"], "Single")
df["Marital_Status"]
```

```python
# K: Plot histogram and box-plot of customers Income and detect outliers visually
df["Income"].plot.hist()
```

```python
df["Income"].plot.box()
```

```python
# L: Remove all outliers with respect to IQR (Inter Quartile Range) and plot cleared data
q1 = df["Income"].quantile(0.25)
q3 = df["Income"].quantile(0.75)
cleared_df = df.loc[df["Income"] < q3]
cleared_df = cleared_df.loc[df["Income"] > q1]
cleared_df["Income"].plot.hist()
```

```python
cleared_df["Income"].plot.box()
```

```python
# M: Draw heat map correlation of dataset
numeric_df = df.select_dtypes(['number'])
numeric_df_corr = numeric_df.corr()
sb.heatmap(numeric_df_corr, 
        xticklabels=numeric_df_corr.columns,
        yticklabels=numeric_df_corr.columns)
```
